```python
df = pd.read_csv(r'C:\Users\hp\Desktop\Jupyter Notebook\Capstone Project EV Sales\New files\all_year.csv')
df.head()
```

```python
# Determine the number of unnamed columns
unnamed_columns = df.columns[df.columns.str.contains('Unnamed', case=False)]
num_unnamed_columns = len(unnamed_columns)

# Determine the indices of the unnamed columns to delete (assuming last 3 columns are unnamed)
unnamed_columns_to_delete = df.columns[-4:]

# Drop the specified unnamed columns
df.drop(columns=unnamed_columns_to_delete, inplace=True)
```

```python
df.head()
```

```python
col = ['Year', 'Sr.no', 'Maker', 'Apr', 'May', "Jun", 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', "Dec", 'Jan', 'Feb', 'Mar', 'Total','MarketShare']
df.columns = col

col
```

```python
df.head()
```

```python
df['Year'] = df['Year'].replace('fy-20', '2019-2020')
df['Year'] = df['Year'].replace('fy-21', '2020-2021')
df['Year'] = df['Year'].replace('fy-21-22', '2021-2022')
df['Year'] = df['Year'].replace('fy-22-23', '2022-2023')
df['Year'] = df['Year'].replace('fy-23-24', '2023-2024')
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Year'].value_counts()
```

```python
# Filter the DataFrame for the fiscal years 2023-2024
df1 = df[df['Year'].str.contains('2023-2024')]
```

```python
df1.head()
```

```python
df1
```

```python
df1 = df1.drop(df1.index[0])
```

```python
df1
```

```python
# Store the 'Feb' and 'Mar' values
feb_values = df1['Feb'].copy()
mar_values = df1['Mar'].copy()



# Shift 'Feb' value to 'Total' column
df1['Total'] = feb_values

# Shift 'Mar' value to 'MarketShare' column
df1['MarketShare'] = mar_values
```

```python
df1
```

```python
# Put NaN values in 'Feb' and 'Mar' columns
df1['Feb'] = float('nan')
df1['Mar'] = float('nan')
```

```python
df1
```

```python

# Drop the last 15 rows
df = df.iloc[:-15]
```

```python
df.tail()
```

```python
import pandas as pd

# Concatenate the two DataFrames vertically
df = pd.concat([df, df1], ignore_index=True)
```

```python
df.tail()
```

```python
df
```

```python
# Filter rows where 'Sr.no' is not equal to 'Total'
df = df[df['Sr.no'] != 'Total']

# Reset index after dropping rows
df.reset_index(drop=True, inplace=True)

```

```python
df
```

```python
# Drop the 'Sr.no' column in place
df.drop(columns=['Sr.no'], inplace=True)
```

```python
df
```

```python
# Drop the last row in place
df.drop(df.index[-1], inplace=True)

```

```python
df
```

```python
# Filter rows where 'Sr.no' is not equal to 'Total'
df = df[df['Maker'] != 'TOTAL']

# Reset index after dropping rows
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df['Maker'] = df['Maker'].str.capitalize()
```

```python
df
```

```python
df['Maker'] = df['Maker'].str.split().str[0]
```

```python
df
```

```python
# Export the DataFrame to a CSV file
df.to_csv('EV_sales_final_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("EV_sales_final_data.csv")
```

```python
df
```

```python
df.dtypes
```

```python
# List of columns to convert
columns_to_convert = [column for column in df.columns if column not in ['Year', 'Maker', 'MarketShare']]

# Remove commas and convert to numeric for each column
for column in columns_to_convert:
    df[column] = df[column].str.replace(',', '').astype(float)

```

```python
df.dtypes
```

```python
# Group by Year and calculate total sales
yearly_sales = df.groupby('Year')['Total'].sum()
print(yearly_sales)
# Plotting
plt.figure(figsize=(10, 7))
yearly_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
